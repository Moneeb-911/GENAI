```python
from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(
    temperature=0,
    model_name='llama-3.1-70b-versatile',
    groq_api_key=''
)
```

```python
import pandas as pd
df = pd.read_csv(
    "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"
)
```

```python
agent = create_pandas_dataframe_agent(model, df, verbose=True,allow_dangerous_code=True)
```

```python
agent.invoke("how many rows are there?")
```

```python
agent.invoke("Give me the count of people survived and died")
```

```python
agent.invoke("the people which survived which person has the lowest age")
```

```python
agent.invoke("return me mean median mode for age")
```
