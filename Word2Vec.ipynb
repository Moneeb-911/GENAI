```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
```

```python
from nltk import sent_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# code to read text file line and by and append it to the list
text=[]
with open('/content/riseAndFallOfRomans.txt','r') as file:
  for line in file:
    text.append(line)
len(text)
```

```python
import nltk
nltk.download('punkt_tab')
text=sent_tokenize(''.join(text))
# performs sentence level tokenization onto the text
```

```python
text[0:10]
```

```python
# performs word level tokenization on each index of text
text=[word_tokenize(sent) for sent in text]
```

```python
text[3]
```

```python
model=Word2Vec(window=10,min_count=2)
```

```python
model.build_vocab(text)
```

```python
model.wv.key_to_index
```

```python
model.train(text,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar("history")
```

```python
model.wv.similarity('history','rome')
```

```python
model.wv.similarity('history','events')
```

```python
vec=model.wv.get_normed_vectors()
```
