```python
# convert whole dataset to lower
df['text']=df['text'].str.lower()
```

```python
# Function to Remove HTML Tags from dataset
import re
def remove_html(s):
    pattern = re.compile(r'<[a-z]* />') 
    return re.sub(pattern,'',s)
```

```python
df['text']=df['text'].apply(remove_html)
```

```python
# function to remove url if necessary
def remove_url(s):
    pattern=re.compile('[w]{3}.[a-z]*.*')
    return re.sub(pattern,'',s)
```

```python
df['text']=df['text'].apply(remove_url)
```

```python
import string
punc=string.punctuation
punc
```

```python
# def remove_punc(s):
#     for i in punc:
#         s=s.replace(i,'')
#     return s
# better function is below in terms of time complexity
def remove_punc(s):
    return s.translate(str.maketrans('','',punc))
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stop_words = list(stopwords.words('english'))
stop_words
```

```python
# function to remove stopwords from the text
def remove_stopwords(s):
    temp=[]
    s=s.split(' ')
    for i in s:
        if i in stop_words:
            continue
        else:
            temp.append(i)
    return " ".join(temp)
```

```python
df['text']=df['text'].apply(remove_stopwords)
```
