```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
import re
import string
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
df=pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
def removeHtmlTags(s):
    pattern=re.compile(r'<[a-z]* />')
    return re.sub(pattern,'',s)
```

```python
df['review']=df['review'].apply(removeHtmlTags)
```

```python
df['review'][1000]
```

```python
df['review']=df['review'].str.lower()
```

```python
# def remove_punc(s):
#     for i in punc:
#         s=s.replace(i,'')
#     return s
# better function is below in terms of time complexity
punc=string.punctuation
def removePunctuations(s):
    return s.translate(str.maketrans('','',punc))
```

```python
df['review']=df['review'].apply(removePunctuations)
```

```python
stop_words=list(stopwords.words('english'))
stop_words
```

```python
def removeStopWords(s):
    lst=s.split(' ')
    res_lst=[]
    for i in lst:
        if i in stop_words:
            continue
        else:
            res_lst.append(i)
    return ' '.join(res_lst)
```

```python
df['review']=df['review'].apply(removeStopWords)
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
import spacy
nlp_model=spacy.load('en_core_web_sm')
```

```python
def lemmatizationWords(s):
    doc_s=nlp_model(s)
    tokens=[token.lemma_ for token in doc_s]
    return ' '.join(tokens)
```

```python
df=df[:10000]
```

```python
df.shape
```

```python
df['review']=df['review'].apply(lemmatizationWords)
```

```python
df['review'][0]
```

```python
sentiment_mapping = {'positive': 0, 'negative': 1}
df['sentiment_num'] = df['sentiment'].map(sentiment_mapping)
```

```python
X_train,X_test,y_train,y_test=train_test_split(df['review'],df['sentiment_num'],test_size=0.2,stratify=df['sentiment_num'])
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X_train)
```

```python
vectorizer.get_feature_names_out()
```

```python
X.shape
```

```python
print(vectorizer.get_feature_names_out())
```

```python
X_test_transformed=vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.metrics import accuracy_score
multi_model=MultinomialNB()
multi_model.fit(X,y_train)
prediction=multi_model.predict(X_test_transformed)
print(accuracy_score(y_test,prediction))
```

```python
gauss_model=GaussianNB()
gauss_model.fit(X.toarray(),y_train)
prediction=gauss_model.predict(X_test_transformed.toarray())
print(accuracy_score(y_test,prediction))
```
